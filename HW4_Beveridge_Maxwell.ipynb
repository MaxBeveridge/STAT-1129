# Question 5

```python
marks = {'Andy':88, 'Amy':66, 'James':90,'Jules':55,'Arthur':77}
t = np.array([])
s = np.array([])
for u,v in marks.items():
    t=np.append(t, v)
    s=np.append(s, u)
print(s,t)
plt.bar(s, t)
plt.title("Student Grades")
plt.show()
plt.pie(t, labels = s)
plt.title("Student Grades")
plt.legend()
plt.show()
```
