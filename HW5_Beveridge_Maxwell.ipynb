# Question 2

```r
Data_Frame <- data.frame (
    id = c(1,2,3,4,5),
    name = c('Peter', 'Amy', 'Ryan', 'Gary', 'Michelle'),
    salary = c(623.30, 515.20, 611.00, 729.00, 843.25))
print(Data_Frame)
```

```r
New_Frame <- cbind(Data_Frame, Department = c('IT', 'Marketing', 'R&D','Sales', 'Management'))
print(New_Frame)
```

```r
New_Frame[c(1,3,5),c(2,3)]
```

```r
x1 <- New_Frame[c(1,4,5),c(2)]
y1 <- New_Frame[c(1,4,5),c(3)]

barplot(y1, names.arg = x1, ylab = 'Salary', xlab = 'Employee', main='Salaries of Employees')
```

```r
x2 <- c(max(New_Frame$salary), min(New_Frame$salary), median(New_Frame$salary))
y2 <- c('Highest Salary', 'Lowest Salary', 'Median Salary')
pie(x2,y2, main = 'Salaries of Employees')
```
