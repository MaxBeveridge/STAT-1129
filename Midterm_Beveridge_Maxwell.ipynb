```python
f = open('numbers.txt', "r")

list = []
newlist = []
for x in f:
    for y in x:
        list.append(y)
for t in range(len(list)):
    if list[t] == ',' or list[t] =='\n':
        list[t]=' '
print(list)
```

```python
newlist = []
for x in range(len(list)):
    if list[x] == '1' and list[x+1]=='0':
        newlist.append(10)
    elif list[x]=='0' and list[x-1]=='1':
        continue
    elif list[x] != ' ':
        newlist.append(int(list[x]))
    else:
        continue


newlist.sort()
print(newlist)
```

```python
freq = {}
for item in newlist:
    if item in freq:
        freq[item]+=1
    else:
        freq[item]=1

for x,y in freq.items():
    print(x, ':', y)
print("Here", end=' ')
for x, y in freq.items():
    print(x, 'occurs', y, 'times', end=', ')
```

```python
import matplotlib.pyplot as plt
x = freq.keys()
y = freq.values()
plt.hist(newlist)
plt.show()
```

```python
import json
with open("./freq.json", "w") as out:
    json.dump(freq, out)
print(json.load(open("freq.json", "r")))
```

```python
import math
import numpy as np
import statistics as stats
```

```python
df = pd.read_csv('amazon-orders.csv')
totals = df["Total Charged"].str.strip("$")
totalsx = totals.tolist()
num = 0
totalsv = []
while num < len(totalsx):
    totalsx[num] = float(totalsx[num])
    totalsv.append(totalsx[num])
    num+=1

a = df.loc[0:9, ["Total Charged"]]
at = a["Total Charged"].str.strip("$")
atx = at.tolist()
num = 0
atv = []
while num < len(atx):
    atx[num] = float(atx[num])
    atv.append(atx[num])
    num+=1
    
m = df.loc[10:27, ["Total Charged"]]
mt = m["Total Charged"].str.strip("$")
mtx = mt.tolist()
num = 0
mtv = []
while num < len(mtx):
    mtx[num] = float(mtx[num])
    mtv.append(mtx[num])
    num+=1

j = df.loc[28:33, ["Total Charged"]]
jt = j["Total Charged"].str.strip("$")
jtx = jt.tolist()
num = 0
jtv = []
while num < len(jtx):
    jtx[num] = float(jtx[num])
    jtv.append(jtx[num])
    num+=1

ju = df.loc[34:58, ["Total Charged"]]
jut = ju["Total Charged"].str.strip("$")
jutx = jut.tolist()
num = 0
jutv = []
while num < len(jutx):
    jutx[num] = float(jutx[num])
    jutv.append(jutx[num])
    num+=1

mean = stats.mean(totalsv)
median = stats.median(totalsv)
averageperday = sum(totalsv)/122
stdev = stats.pstdev(totalsv)
sumt = sum(totalsv)
mint = min(totalsv)
maxt = max(totalsv)
amean = stats.mean(atv)
mmean = stats.mean(mtv)
jmean = stats.mean(jtv)
jumean = stats.mean(jutv)

print("A total of $", sumt, "was spent over the course of four months.")
print("The average purchase was $", mean)
print("The average amount of money spent per day was $", averageperday)
print("The standard deviation of the total charged was $", stdev)
print("The median total amount charged was $", median)
print("The cheapest purchase was $", mint)
print("The most expensive purchase was $", maxt)
print("The average purchase in April was $", amean)
print("The average purchase in May was $", mmean)
print("The average purchase in June was $", jmean)
print("The average purchase in July was $", jumean)
```
